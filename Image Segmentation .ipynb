# Semantic Segmentation

1. Import the necessary libaries.

```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import ndimage
```

2. Load the image

```python
image = plt.imread('test.jpg')
image.shape
plt.imshow(image)
```

3. Convert the image color to gray 

```python
c_gray = rgb2gray(image)
plt.imshow(c_gray, cmap='gray')
```

4. Compute the shape of the  gray image

```python
c_gray.shape
```

5. Take the mean of the pixel value and use it as threshold . If , pixel value>threshold -> it belongs to the image, else ->it belongs to the background.

```python
gray_new = c_gray.reshape(c_gray.shape[0]*c_gray.shape[1])
for i in range(gray_new.shape[0]):
    if gray_new[i] > gray_new.mean():
        gray_new[i] = 1
    else:
        gray_new[i] = 0
gray = gray_new.reshape(c_gray.shape[0],c_gray.shape[1])
plt.imshow(gray, cmap='gray')
```
